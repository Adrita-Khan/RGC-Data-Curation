```python
import numpy as np

# Load labels
labels = np.load("labels_pred.npy")

# Basic stats
unique, counts = np.unique(labels, return_counts=True)
total = len(labels)

print("Label Statistics:")
for label, count in zip(unique, counts):
    print(f"Class {label}: {count} samples ({(count / total) * 100:.2f}%)")

print("\nSummary:")
print(f"Total samples: {total}")
print(f"Unique classes: {len(unique)}")
print(f"Class labels: {unique}")

# Identify missing labels from the expected full range
expected_range = set(range(703))  # 0 to 702 inclusive
present_labels = set(unique)
missing_labels = sorted(expected_range - present_labels)

print(f"\nMissing Labels ({len(missing_labels)}):")
print(missing_labels)
```
